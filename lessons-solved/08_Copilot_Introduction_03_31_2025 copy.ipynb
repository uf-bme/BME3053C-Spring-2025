- Suggesting whole lines or blocks of code as you type

# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    # Input weight and height from the user
    try:
        weight = float(input("Enter your weight in kg: "))
        height = float(input("Enter your height in meters: "))
        
        # Calculate BMI
        bmi = calculate_bmi(weight, height)
        
        # Determine BMI category
        category = bmi_category(bmi)
        
        # Output the results
        print(f"Your BMI is: {bmi}")
        print(f"You are classified as: {category}")
    except ValueError as e:
        print(e)
if __name__ == "__main__":
    main()
    
    
````

## 3. Working with Libraries

Copilot is particularly helpful when working with libraries. It can suggest proper import statements and usage patterns:

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

# Let Copilot suggest the implementation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate random data points
np.random.seed(0)
x = np.random.rand(100) * 10
y = 2 * x + np.random.randn(100) * 5  # Adding some noise for realism

# Assign custom colors based on y values
colors = y - y.min()  # Normalize values for color mapping

# Create a scatter plot
plt.scatter(x, y, c=colors, cmap='viridis', label='Data Points')

degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
polyreg.fit(x.reshape(-1, 1), y)

# Generate trend line
x_fit = np.linspace(0, 10, 100)
y_fit = polyreg.predict(x_fit.reshape(-1, 1))

# Plot trend line
plt.plot(x_fit, y_fit, color='red', label=f'Trend Line (Degree {degree})')

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot with Trend Line and Custom Colors')

# Add legend
plt.legend()
# Show the plot
plt.colorbar(label='Color Scale (Based on Y values)')
plt.show()


````

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Processes a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    try:
        # Validate input
        if not isinstance(image, np.ndarray):
            raise ValueError("Input must be a numpy array.")
        if image.size == 0:
            raise ValueError("Input image is empty.")

        # Normalize the image (scale values to range [0, 1])
        normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Apply a Gaussian filter for smoothing
        filtered_image = gaussian_filter(normalized_image, sigma=1)

        return filtered_image

    except Exception as e:
        print(f"Error processing image: {e}")
        raise

    
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    def __init__(self):
        """Initialize an empty dictionary to store patient records."""
        self.records = {}

    def add_record(self, patient_id, data):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            data (dict): Patient data (e.g., name, age, medical history).

        Raises:
            ValueError: If patient_id already exists or data is invalid.
        """
        if patient_id in self.records:
            raise ValueError(f"Patient ID {patient_id} already exists.")
        if not isinstance(data, dict) or 'name' not in data or 'age' not in data:
            raise ValueError("Invalid data. Must include 'name' and 'age'.")
        self.records[patient_id] = data

    def update_record(self, patient_id, data):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            data (dict): Updated patient data.

        Raises:
            ValueError: If patient_id does not exist or data is invalid.
        """
        if patient_id not in self.records:
            raise ValueError(f"Patient ID {patient_id} does not exist.")
        if not isinstance(data, dict):
            raise ValueError("Invalid data. Must be a dictionary.")
        self.records[patient_id].update(data)

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient data.

        Raises:
            ValueError: If patient_id does not exist.
        """
        if patient_id not in self.records:
            raise ValueError(f"Patient ID {patient_id} does not exist.")
        return self.records[patient_id]

    def get_average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: Average age of patients.

        Raises:
            ValueError: If no records exist.
        """
        if not self.records:
            raise ValueError("No patient records available.")
        ages = [data['age'] for data in self.records.values() if 'age' in data]
        return sum(ages) / len(ages)

    def get_patient_count(self):
        """
        Get the total number of patients.

        Returns:
            int: Number of patients.
        """
        return len(self.records)
    
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets

# Generate sample data
np.random.seed(0)
data = {
    "Category": np.random.choice(["A", "B", "C"], 100),
    "Value1": np.random.randn(100) * 10 + 50,
    "Value2": np.random.randn(100) * 5 + 20,
    "Date": pd.date_range("2025-01-01", periods=100)
}
df = pd.DataFrame(data)

# Function to display a histogram
def plot_histogram(column):
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=15, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Function to display a line plot
def plot_line(category):
    filtered_data = df[df["Category"] == category]
    plt.figure(figsize=(8, 5))
    plt.plot(filtered_data["Date"], filtered_data["Value1"], label="Value1", marker='o')
    plt.plot(filtered_data["Date"], filtered_data["Value2"], label="Value2", marker='x')
    plt.title(f"Line Plot for Category {category}")
    plt.xlabel("Date")
    plt.ylabel("Values")
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to display a scatter plot
def plot_scatter(x_column, y_column):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x_column], df[y_column], c='green', alpha=0.7)
    plt.title(f"Scatter Plot: {x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.show()

# Interactive widgets
print("Interactive Dashboard:")
interact(plot_histogram, column=widgets.Dropdown(options=["Value1", "Value2"], description="Histogram:"))
interact(plot_line, category=widgets.Dropdown(options=df["Category"].unique(), description="Line Plot:"))
interact(plot_scatter, 
         x_column=widgets.Dropdown(options=["Value1", "Value2"], description="X-axis:"),
         y_column=widgets.Dropdown(options=["Value1", "Value2"], description="Y-axis:"))
````
