# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
    
def main():
    # Input weight and height from the user
    try:
        weight = float(input("Enter your weight in kg: "))
        height = float(input("Enter your height in meters: "))
        
        # Calculate BMI
        bmi = calculate_bmi(weight, height)
        
        # Get BMI category
        category = bmi_category(bmi)
        
        # Output the results
        print(f"Your BMI is: {bmi}")
        print(f"You are classified as: {category}")
    except ValueError as e:
        print(e)

main()
    
````

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Generate random data
np.random.seed(42)  # for reproducibility
x = np.random.rand(50)
y = 0.5 * x + np.random.normal(0, 0.1, 50)  # Create correlated data

# Calculate trend line
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# Create color map based on y values
colors = y

# Create the plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, cmap='viridis', 
                     alpha=0.6, label='Data points')
plt.plot(x, line, color='red', label=f'Trend line (R짼 = {r_value**2:.3f})')

# Customize the plot
plt.title('Random Data with Trend Line', fontsize=12, pad=15)
plt.xlabel('X Values', fontsize=10)
plt.ylabel('Y Values', fontsize=10)

# Add color bar
plt.colorbar(scatter, label='Y Value Magnitude')

# Add legend
plt.legend(loc='upper left')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

plt.show()

````

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from ipywidgets import interact, FloatSlider
%matplotlib inline

def plot_with_slope(slope_factor):
    # Generate random data
    np.random.seed(42)  # for reproducibility
    x = np.random.rand(50)
    y = slope_factor * x + np.random.normal(0, 0.1, 50)  # Adjustable slope
    
    # Calculate trend line
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    line = slope * x + intercept
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(x, y, c=y, cmap='viridis', 
                         alpha=0.6, label='Data points')
    plt.plot(x, line, color='red', label=f'Trend line (R짼 = {r_value**2:.3f})')
    
    # Customize the plot
    plt.title(f'Random Data with Trend Line (Slope Factor: {slope_factor:.2f})', 
             fontsize=12, pad=15)
    plt.xlabel('X Values', fontsize=10)
    plt.ylabel('Y Values', fontsize=10)
    plt.ylim(-5, 5)  # Set y-axis limits
    
    # Add color bar
    plt.colorbar(scatter, label='Y Value Magnitude')
    
    # Add legend
    plt.legend(loc='upper left')
    
    # Add grid
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Adjust layout
    plt.tight_layout()
    plt.show()

# Create interactive slider
interact(plot_with_slope, 
    slope_factor=FloatSlider(
        value=0.5,
        min=-5.0,
        max=5.0,
        step=0.1,
        description='Slope:',
        continuous_update=False
    ))
````

````python
def calculate_dosage(weight, age, condition):
    """Calculate medication dosage based on patient parameters."""
    # Validate that weight and age are positive values
    if weight <= 0 or age < 0:
        raise ValueError("Weight and age must be positive")
    
    # Calculate the base dosage as twice the weight
    base_dosage = weight * 2
    
    # Adjust dosage for children under 12 years old
    if age < 12:
        base_dosage *= 0.75
    
    # Increase dosage by 50% for severe conditions
    if condition == "severe":
        base_dosage *= 1.5
    
    # Return the final dosage rounded to 2 decimal places
    return round(base_dosage, 2)

# Import ipytest for running test cases in Jupyter Notebook
import ipytest

# Define pytest test cases for the calculate_dosage function
def test_calculate_dosage():
    # Test normal case for an adult
    assert calculate_dosage(70, 30, "normal") == 140.0
    # Test normal case for a child under 12
    assert calculate_dosage(50, 10, "normal") == 75.0
    # Test severe condition for a teenager
    assert calculate_dosage(80, 15, "severe") == 240.0
    # Test normal case for a young child
    assert calculate_dosage(60, 5, "normal") == 90.0

# Run the tests
ipytest.run("-v")
````

### 截 Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Process a medical image represented as a numpy array.
    
    Parameters:
        image (numpy.ndarray): Input image array.
    
    Returns:
        numpy.ndarray: Processed image array.
    
    Raises:
        ValueError: If the input is not a valid numpy array.
    """
    # Error handling: Check if the input is a valid numpy array
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    
    if image.ndim != 2:
        raise ValueError("Input image must be a 2D array.")
    
    try:
        # Step 1: Normalize the image (scale pixel values to range [0, 1])
        normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))
        
        # Step 2: Apply Gaussian filtering for noise reduction
        filtered_image = gaussian_filter(normalized_image, sigma=1)
        
        # Return the processed image
        return filtered_image
    
    except Exception as e:
        raise RuntimeError(f"An error occurred during image processing: {e}")

# Example usage of the process_medical_image function
# Create a synthetic medical image (2D numpy array)
image = np.random.rand(256, 256) * 255  # Random grayscale image
# Process the image using the defined function
processed_image = process_medical_image(image)
# Display the raw and processed images as two subplots
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Raw image subplot
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Raw Medical Image")
plt.colorbar()

# Processed image subplot
plt.subplot(1, 2, 2)
plt.imshow(processed_image, cmap='gray')
plt.title("Processed Medical Image")
plt.colorbar()

plt.tight_layout()
plt.show()
````


### 截 Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records, including adding, updating, retrieving,
    and performing basic statistical analysis.
    """
    def __init__(self):
        # Initialize an empty dictionary to store patient records
        self.records = {}

    def add_record(self, patient_id, name, age, weight, condition):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            weight (float): Weight of the patient in kilograms.
            condition (str): Medical condition of the patient.

        Raises:
            ValueError: If any input data is invalid.
        """
        if not isinstance(patient_id, str) or not patient_id:
            raise ValueError("Patient ID must be a non-empty string.")
        if not isinstance(name, str) or not name:
            raise ValueError("Name must be a non-empty string.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if not isinstance(weight, (int, float)) or weight <= 0:
            raise ValueError("Weight must be a positive number.")
        if not isinstance(condition, str) or not condition:
            raise ValueError("Condition must be a non-empty string.")

        self.records[patient_id] = {
            "name": name,
            "age": age,
            "weight": weight,
            "condition": condition
        }

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            kwargs: Key-value pairs of fields to update.

        Raises:
            KeyError: If the patient ID does not exist.
            ValueError: If any input data is invalid.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")

        for key, value in kwargs.items():
            if key not in self.records[patient_id]:
                raise ValueError(f"Invalid field: {key}")
            if key == "age" and (not isinstance(value, int) or value <= 0):
                raise ValueError("Age must be a positive integer.")
            if key == "weight" and (not isinstance(value, (int, float)) or value <= 0):
                raise ValueError("Weight must be a positive number.")
            self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: The patient's record.

        Raises:
            KeyError: If the patient ID does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")
        return self.records[patient_id]

    def get_average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: The average age.

        Raises:
            ValueError: If no records exist.
        """
        if not self.records:
            raise ValueError("No records available to calculate average age.")
        return sum(record["age"] for record in self.records.values()) / len(self.records)

    def get_average_weight(self):
        """
        Calculate the average weight of all patients.

        Returns:
            float: The average weight.

        Raises:
            ValueError: If no records exist.
        """
        if not self.records:
            raise ValueError("No records available to calculate average weight.")
        return sum(record["weight"] for record in self.records.values()) / len(self.records)

    def get_patient_count(self):
        """
        Get the total number of patients.

        Returns:
            int: The number of patients.
        """
        return len(self.records)
    
# Example usage of the PatientRecords class
patient_records = PatientRecords()
# Adding records
patient_records.add_record("P001", "John Doe", 30, 70.0, "Healthy")
patient_records.add_record("P002", "Jane Smith", 25, 60.0, "Healthy")
patient_records.add_record("P003", "Alice Johnson", 40, 80.0, "Diabetic")
# Printing all patient records
for patient_id, record in patient_records.records.items():
    print(f"Patient ID: {patient_id}, Record: {record}")
# Updating a record
patient_records.update_record("P002", weight=62.5)
# Retrieving a record
record = patient_records.get_record("P002")
print(f"Record for P002: {record}")
# Getting average age and weight
average_age = patient_records.get_average_age()
average_weight = patient_records.get_average_weight()
print(f"Average Age: {average_age}")
print(f"Average Weight: {average_weight}")
# Getting total patient count
patient_count = patient_records.get_patient_count()
print(f"Total Patients: {patient_count}")
# Example of error handling
try:
    patient_records.update_record("P999", weight=65.0)  # Non-existent ID
except KeyError as e:
    print(e)
# Example of error handling for invalid input
try:
    patient_records.add_record("", "Invalid Patient", -5, -70.0, "")
except ValueError as e:
    print(e)
# Example of error handling for empty records
try:
    empty_records = PatientRecords()
    empty_records.get_average_age()
except ValueError as e: 
    print(e)
# Example of error handling for invalid field update

````

### 截 Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown, IntSlider

# Sample dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 45],
    "Weight": [55, 70, 68, 80, 60],
    "Height": [160, 175, 170, 180, 165],
    "Condition": ["Healthy", "Diabetic", "Healthy", "Hypertensive", "Healthy"]
}
df = pd.DataFrame(data)

# Function to display a bar plot
def plot_bar(column):
    plt.figure(figsize=(8, 5))
    df[column].value_counts().plot(kind="bar", color="skyblue")
    plt.title(f"Bar Plot of {column}", fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Function to display a scatter plot
def plot_scatter(x_column, y_column):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x_column], df[y_column], color="green", alpha=0.7)
    plt.title(f"Scatter Plot: {x_column} vs {y_column}", fontsize=14)
    plt.xlabel(x_column, fontsize=12)
    plt.ylabel(y_column, fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()

# Function to filter data by age and display a summary
def filter_data(min_age, max_age):
    filtered_df = df[(df["Age"] >= min_age) & (df["Age"] <= max_age)]
    print(f"Filtered Data (Age {min_age}-{max_age}):")
    display(filtered_df)

# Interactive bar plot
print("Interactive Bar Plot:")
interact(plot_bar, column=Dropdown(options=df.columns, description="Column:"))

# Interactive scatter plot
print("\nInteractive Scatter Plot:")
interact(plot_scatter, 
         x_column=Dropdown(options=df.columns, description="X-axis:"),
         y_column=Dropdown(options=df.columns, description="Y-axis:"))

# Interactive data filter
print("\nInteractive Data Filter:")
interact(filter_data, 
         min_age=IntSlider(value=20, min=20, max=50, step=5, description="Min Age:"),
         max_age=IntSlider(value=50, min=20, max=50, step=5, description="Max Age:"))
````
